2. Замените в массиве все нечетные положительные числа на -1

```python
array_a = np.random.randint(-10, 10, 10)
print(array_a)
array_a[(array_a > 0) & (array_a % 2 !=0)] = -1
print(array_a)
```

3. Найдите все числа, встречающиеся в двух массивах

```python
array_a = np.random.randint(-10, 10, 10)
array_b = np.random.randint(-10, 10, 10)
print(array_a)
print(array_b)
print(np.intersect1d(array_a,array_b))
```

5. В данном массиве замените все числа, меньшие 0 на 0, а большие 255 на 255. Напишите минимум два варианта решения

```python
array_b = np.random.randint(-300, 300, 10)
print(array_b)
array_b = np.where(array_b < 0, 0, array_b)
array_b = np.where(array_b > 255, 255, array_b)
array_b

array_c = np.random.randint(-300, 300, 10)
print(array_c)
array_c[array_c < 0] = 0
array_c[array_c > 255] = 255
array_c

array_d = np.random.randint(-300, 300, 10)
print(array_d)
array_d = np.clip(array_d, 0, 255)
array_d
```

8. Удалите все np.nan из массива. 
*Для того, чтобы в тестовом массиве появился nan используйте np.nan или деление 0. на 0.*

```python
array_i = np.array([1, 2, 3, np.nan, 5, 6, np.nan, 8, 9, np.nan])

array_i = array_i[np.logical_not(np.isnan(array_i))]
array_i
```
